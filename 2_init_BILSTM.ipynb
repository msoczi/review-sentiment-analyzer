# 2. Initialization of BILSTM model

To classify the sentiment of a given review, we will first use the bidirectional LSTM (BiLSTM) model. The neural networks were created using `tensorflow`.

For the data prepared in the previous script, we initiated a BILSTM network. The network architecture is very simple. The network consists of only two layers: word embedding and the bidirectional LSTM layer. Then trained one epoch - 40 minuts on GPU with Google Colab.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import numpy as np
import pandas as pd
import pickle

import tensorflow as tf
import keras
from keras.models import Sequential
from keras import layers
from tensorflow.keras.optimizers import RMSprop, Adam
```

```python
# Read the data
X_train = np.load('/content/drive/MyDrive/INL_PROJEKT/preprocessed_data/X_train.npy')
X_test = np.load('/content/drive/MyDrive/INL_PROJEKT/preprocessed_data/X_test.npy')
y_train = np.load('/content/drive/MyDrive/INL_PROJEKT/preprocessed_data/y_train.npy')
y_test = np.load('/content/drive/MyDrive/INL_PROJEKT/preprocessed_data/y_test.npy')

# Read tokenizer
with open('/content/drive/MyDrive/INL_PROJEKT/preprocessed_data/tokenizer.pickle', 'rb') as f:
    tokenizer = pickle.load(f)
```

```python
max_words = 5000
max_len = 600

# Create the BILSTM model for sentiment analysis
model_BILSTM = Sequential()
model_BILSTM.add(layers.Embedding(max_words, 30, input_length=max_len))  # 40
model_BILSTM.add(layers.Bidirectional(layers.LSTM(15,dropout=0.6)))  # LSTM(20, dropuot=0.6)
model_BILSTM.add(layers.Dense(3,activation='softmax'))
model_BILSTM.compile(optimizer='rmsprop',loss='categorical_crossentropy', metrics=['accuracy'])

# Architecture of model
model_BILSTM.summary()
```

```python
# Train BILSTM with one epoch
history = model_BILSTM.fit(X_train, 
                     y_train, 
                     epochs=1,
                     validation_data=(X_test, y_test),
                     verbose = 1)
```

```python
# Save the model
model_BILSTM.save('/content/drive/MyDrive/INL_PROJEKT/BILSTM/model_epoch_1')
```
